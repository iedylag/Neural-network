```python
from sklearn.neural_network import MLPClassifier
import numpy as np

def toArray(string):
    if len(string) != 7 * 7:
        raise ValueError('string in wrong size')
    return np.array([toNumber(char) for char in string])

def toNumber(character):
    return 1 if character == '#' else 0

zero = toArray(
    '#######' +
    '#     #' +
    '#     #' +
    '#     #' +
    '#     #' +
    '#     #' +
    '#######'
)
one = toArray(
    '   #   ' +
    '   #   ' +
    '   #   ' +
    '   #   ' +
    '   #   ' +
    '   #   ' +
    '   #   '
)
two = toArray(
    '#######' +
    '#     #' +
    '      #' +
    '     # ' +
    '   #   ' +
    ' #     ' +
    '#######'
)
three = toArray(
    '#######' +
    '      #' +
    '      #' +
    ' ######' +
    '      #' +
    '      #' +
    '#######'
)
four = toArray(
    '#     #' +
    '#     #' +
    '#     #' +
    '#######' +
    '      #' +
    '      #' +
    '      #'
)
five = toArray(
    '#######' +
    '#      ' +
    '#      ' +
    '#######' +
    '      #' +
    '      #' +
    '#######'
)
six = toArray(
    '      #' +
    '    #  ' +
    '  #    ' +
    ' ######' +
    '#     #' +
    '#     #' +
    '#######'
)
seven = toArray(
    '#######' +
    '     # ' +
    '    #  ' +
    '   #   ' +
    '  #    ' +
    ' #     ' +
    '#      '
)
eight = toArray(
    '#######' +
    '#     #' +
    '#     #' +
    '#######' +
    '#     #' +
    '#     #' +
    '#######'
)
nine = toArray(
    '#######' +
    '#     #' +
    '#     #' +
    '###### ' +
    '    #  ' +
    '   #   ' +
    ' #     '
)

X = np.array([zero, one, two, three, four, five, six, seven, eight, nine])
y = np.array(['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine'])

net = MLPClassifier(hidden_layer_sizes=(100,), activation='logistic', solver='lbfgs', max_iter=500)
net.fit(X, y)

# result = net.predict([toArray(
#     '#######' +
#     '#     #' +
#     '#     #' +
#     '##  ###' +
#     '#     #' +
#     '#     #' +
#     '#######'
# )])

# result = net.predict([toArray(
#     '#######' +
#     '     # ' +
#     '    #  ' +
#     '   #   ' +
#     '   #   ' +
#     '   #   ' +
#     '   #   '
# )])

result = net.predict([toArray(
    '     # ' +
    '     # ' +
    '     # ' +
    '     # ' +
    '     # ' +
    '     # ' +
    '     # '
)])

print(result)
```
